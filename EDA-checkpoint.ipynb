```python
# data is messy needs cleaning => start cleaning
#.. check for null values
data.isnull().sum()
```

```python
# remove unwanted column(s) 
England = data.drop(['Unnamed: 6',], axis=1)
```

```python
England.columns
```

```python
# remove rows where there are null values
England.dropna(how = 'all', inplace = True)
```
